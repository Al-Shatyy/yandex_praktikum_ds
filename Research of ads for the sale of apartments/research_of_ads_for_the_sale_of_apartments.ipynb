## Предобработка данных

### Обработка пропусков


#### *floors_total*

```python
print('Пропущенных значений в floors_total:', data['floors_total'].isna().sum())
```

Пропущенные значения можно востановить с высокой точностью, зная расстояние до центра *cityCenters_nearest* и вторую точку, например расстояние до аэропорта *airports_nearest*. Квартиры, находящиеся на одинаковом удалении от двух точек, вероятнее всего, находятся в одном доме. Сгруппируем таблице по *cityCenters_nearest* и *airports_nearest*, посчитаем среднюю этажность и заполним этим значением пропуски.

```python
data['floors_total'].fillna(
    data.groupby(['cityCenters_nearest', 'airports_nearest'])['floors_total'].transform('mean'),
    inplace=True)
```

```python
print(data['floors_total'].isna().sum(), '- число оставшихся пропусков')
```

```python
data['floors_total'].fillna(-1, inplace=True)
```


#### *is_apartment*

```python
data['is_apartment'].value_counts(dropna=False)
```

### Замена типа данных

Заменим тип данных 
- *last_price, floors_total, balcony* на int;
- *first_day_exposition* на datetime;
- *is_apartment* на bool.

```python
data['last_price'] = data['last_price'].astype('int')
data['floors_total'] = data['floors_total'].round().astype('Int64')
data['balcony'] = data['balcony'].astype('int')

data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d')

data['is_apartment'] = data['is_apartment'].astype('bool')
```

#### *locality_name*

Сначала заполним пропуски значений словом "Неизвестно".

```python
data['locality_name'].fillna('Неизвестно', inplace=True)
```

```python
locality_unique = data['locality_name'].unique()
print('Количество уникальных названий:', len(locality_unique))
```

Мы имеем 365 уникальных названий населенных пунктов. Рассмотрим их детальнее.

```python
locality_unique[:50]
```

Названия населенных пунктов содержат определения типа населенного пункта (деревня, поселок...). Также есть записи различающиеся только наличием типа населенного пункта (посёлок Мурино, Мурино), очевидно это один населенный пункт.

```python
locality_splitted = [locality.split() for locality in locality_unique]
```

```python
counter = Counter()
for locality in locality_splitted:
    counter += Counter(locality)
    
counter.most_common(15)
```

Мы получили данные по частоте употребления отдельных слов в столбце *locality_name*. Среди них можно выделить наиболее употребимые слова, которые и являются обозначениями типа населенного пункта. Выделим эти слова в отдельное множество.

```python
locality_types = set(['посёлок', 'поселок', 'деревня', 'городского', 'типа', 'городской', 
                      'село', 'товарищество', 'садовое', 'коттеджный'])
```

Используя это множество исключим из значений местности названия типов населенных пунктов.

```python
data['locality_name'] = [' '.join(list(set(locality.split()) - locality_types)) 
                         for locality in data['locality_name'].values]
```

```python
print('Количество уникальных названий:', len(data['locality_name'].unique()))
```

Число уникальных названий сократилось с 365 до 309. Это облегчит дальнейший анализ и поможет увеличить его точность.

## Расчёты и добавление результатов в таблицу

### Цена квадратного метра

Создадим столбцец *price_per_square* как результат деления цены квартиры на ее общую площадь.

```python
data['price_per_square'] = (data['last_price'] // data['total_area']).astype('int')
data.head()
```

### День недели, месяц и год публикации

Добавим столбцы со значением дня недели, месяца и года публикации каждого объявления. Исходные данные возьмем из столбца *first_day_exposition*.

```python
data['exposition_weekday'] = data['first_day_exposition'].dt.weekday
data['exposition_month'] = data['first_day_exposition'].dt.month
data['exposition_year'] = data['first_day_exposition'].dt.year

data.head()
```

### Этаж квартиры

Напишем функцию, которая будет сравнивать значения столбцов *floor* и *floors_total*. Если значение этажа равно 1, но возвращается строка 'первый', если значения этажа и общего числа этажей в доме равны, то возвращается строка 'последний', в остальных случаях возвращается строка 'другое'.

```python
def floors_type(row):
    floors_total = row['floors_total']
    floor = row['floor']
    
    if floors_total > 0:
        if floor == 1:
            return 'первый'
        elif floor == floors_total:
            return 'последний'
    return 'другое'
```

```python
data['floor_type'] = data.apply(floors_type, axis=1)

data.head()
```

### Соотношение площадей квартиры

Добавим столбец *living_area_propotion*, показывающий соотношение жилой площади квартиры к общей площади.

```python
data['living_area_propotion'] = data['living_area'] / data['total_area']
```

Также добавим столбец *kitchen_area_propotion*, показывающий соотношение площадей кухни к общему.

```python
data['kitchen_area_propotion'] = data['kitchen_area'] / data['total_area']
```

Используя эти соотношения можно попробовать восстановить пропуски в столбцах *kitchen_area*, *living_area*. Посчитаем средние значения соотношений для групп объединенных по общей площади и числу комнат. По средним значениям соотношений площадей, общей площади и числу комнат объектов восстановим жилую площадь и площадь кухни.

```python
df = data.copy()
df['total_area'] = df['total_area'].round()
grouped = df.groupby(['total_area','rooms'])
```

```python
for column in ['living_area', 'kitchen_area']:
    data[column] = data[column].fillna(
        grouped[column].transform(
            lambda x: x.mean() / (x.name[0])) * data['total_area']).round(2)
```

В столбце присутствуют явные выбросы (максимальное значение 100 метров и минимальное - 1 метр). Изучим детальнее верхний и нижний пределы значений столбца.

```python
sorted(data.query('ceiling_height > 6').ceiling_height.unique())
```

Самое большое значение *ceiling_height* равно 100. Это значение лучше удалить, так как оно является выбросом.

```python
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height'] < 100)
```

В списке значений *ceiling_height* заметна группа аномальных значений (от 20.0 до 32.0). Возможная причина появления таких значений - смещение десятичного разделителя вправо. Преобразуем аномальные значения.

```python
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height'] < 20, data['ceiling_height'] / 10)
```

Проверим нижнюю границу значений.

```python
sorted(data.query('ceiling_height < 2.5').ceiling_height.unique())
```

Значения меньше 2.0 - слишком низкие для высоты потолков квартиры. Удалим эти значения. Также округлим значения до двух цифр после разделителя. 

Здесь важно учитывать особенности реальных данных. Если удалять крайние и необычные значения, то в дальнейшем, когда эти данные будут использоваться для обучения моделей машинного обучения, модель не сможет полноценно работать. Когда модели при работе с реальными данными потребуется определить стоимость жилья с необычными параметрами, то результат может оказаться некорректным.

```python
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height'] >= 2.0).round(2)
```

Проверим изменения в данных.

```python
print(data['ceiling_height'].describe())
```

Высота потолков в квартирах имеет выраженные дискретные значения, такие как 2.5, 2.6, 2.7 и 3.0. Среднее значение высоты потолков - 2.7 м., медианное - 2,65 м. В результате изменений среднее и медианное значения не изменились.
