### Обработка отсутствующих значений

Удалим строки с отсутствующими значениями названия и даты выхода игры.

```python
data = data.dropna(subset=['name', 'genre']).reset_index(drop=True)
```

```python
data.isna().sum()
```

Заполним пропущенные значения в столбце *year_of_release*.

Создадим словарь, где соотнесем каждую платформу и средний год выхода игр на этой платформе.

```python
release_date_by_platform = data.groupby('platform')['year_of_release'].mean().astype('int').to_dict()
```

Заполним пропущенные значения даты выхода игр наиболее вероятными значениями. Функция *fillna_year_of_release* находит предполагаемый год из названия игры. Если поиск по названию игры не дает результата, то возвращается среднее значение года по соответствующей консоли.

```python
def fillna_year_of_release(row):
    name = row['name']
    year_of_release = row['year_of_release']
    platform = row['platform']
    expression = r'\b\d{4}\b'
    
    if year_of_release > 0:
        return year_of_release
    
    else:
        reg_val = search(expression, name)
        if reg_val:
            result = int(reg_val.group(0))
            if min_val <= result <= max_val:
                return result
        return release_date_by_platform[platform]
```

```python
min_val, max_val = data['year_of_release'].min(), data['year_of_release'].max()
data['year_of_release'] = data.apply(fillna_year_of_release, axis=1)
```

Заполнены все пропуски в столбце *year_of_release*, найденные значения не выходят за пределы распределения для *year_of_release*.

```python
data['year_of_release'].describe()
```

Среди значений *user_score* имеется значене 'tbd'. Эта аббревиатура может означать 'To Be Determined' - 'будет определено позже'. Количество записей с "tbd" составляет 14.5% об общего числа.

### Продажи по платформам с распределением по годам

Посчитаем объем продаж игр для каждой платформы и определим **топ платформ** по продажам.

```python
sales_by_platform = data.groupby('platform')['total_sales'].sum().sort_values(ascending=False)

sales_by_platform.describe()
```

Выделим платформы с наибольшими продажами как платформы, у которых **продажи больше медианы по всей выборке**.

```python
top_platforms = sales_by_platform[sales_by_platform > sales_by_platform.median()]
display(top_platforms)

top_platforms = top_platforms.index
```

Составим сводную таблицу для каждой платформы из топа со значениями количества вышедших игр по годам. По полученным данным построим диаграмму.

```python
top_platform_sales = (data[data['platform'].isin(top_platforms)]
                     .pivot_table(
                         columns='platform',
                         index='year_of_release',
                         values='name',
                         aggfunc='count'))
```

```python
plt.figure(figsize=(19,16))

sns.heatmap(top_platform_sales, linewidths=.5, cmap='YlOrRd', annot=True, fmt='2.0f')
plt.title('Количество новых игр для самых популярных платформ по годам (шт.)')
plt.yticks(rotation=0)
plt.show()
```

Удалим аномальное значение для платформы DS в 1985 году.

```python
data = data.drop(data.query('platform=="DS" & year_of_release==1985').index)
```

По диаграмме можно проследить для каждой платформы отрезок времени, когда разработчики выпускали для нее новые игры. Посчитаем среднее "время жизни" платформ, исключив из выборки те платформы, для которых по данным на 2016 год активно выпускают игры. 

```python
platforms_2016 = top_platform_sales.loc[[2016]]
recent_sales_platform = platforms_2016.sum().where(lambda x: x > 10).dropna().index
```

```python
mean_lifetime = top_platform_sales.drop(recent_sales_platform, axis=1).count().mean()
print('Среднее время "жизни" платформ: {:.1f} лет'.format(mean_lifetime))
```

### Определение актуального периода

Выделим актуальные платформы как те, на которые в 2016 году выпускали новые игры. 

```python
actual_platforms = data.query('year_of_release == 2016')['platform'].unique()

actual_platforms
```

```python
actual_releases = data.loc[(data['platform'].isin(actual_platforms)) & (data['year_of_release']>2000)]
actual_platform_releases = (actual_releases
                            .pivot_table(
                                columns='platform',
                                index='year_of_release',
                                values='name',
                                aggfunc='count'))
```

Построим диграмму выхода новых игр для актуальных платформ по годам.

```python
plt.figure(figsize=(16, 9))

sns.heatmap(actual_platform_releases, linewidths=.5, cmap='YlOrRd', annot=True, fmt='2.0f')
plt.title('Количество новых игр для актуальных платформ по годам (шт.)')
plt.yticks(rotation=0)
plt.show()
```

```python
score_push_sales = actual_data[['platform', 'total_sales', 'user_score', 'critic_score']]
```

Построим матрицу рассеяния для данных по платформа PS4 и посчитаем коэффициенты корреляции.

```python
sns.set_style('whitegrid')

plt.figure(figsize=(10, 10))

pd.plotting.scatter_matrix(
    score_push_sales[score_push_sales['platform']=='PS4'],
    figsize=(9, 9),
    alpha=0.6,
    color=(0.4, 0.5, 1, 1))

plt.suptitle("PS4")
plt.show()
```

На диаграммах видно, что **большая часть хорошо продаваемых игр имеют высокую оценку как пользователей, так и критиков**. Пользователи более склонны оценивать популярные игры низко, чем критики - диаграмма зависимости продаж от оценок пользователей оказалась более рассеянной. Возможно эта особенность стала причиной того, что коэффициент корреляции продаж и оценок пользователей практически равен нулю.

Корреляция продаж и оценок критиков более выраженная, положительная. Также наблюдается средняя корреляция между оценками пользователей и критиков.

```python
def plot_corr(platform):
    console_data = score_push_sales[score_push_sales['platform']==platform]
    ylim = ceil(console_data['total_sales'].quantile(q=0.98))
    
    fig, axes = plt.subplots(1, 2, figsize=(13, 5))
    sns.scatterplot(data=console_data, ax=axes[0], x='user_score', y='total_sales', alpha=0.7, s=49, color=(0.4, 0.5, 1, 1))
    axes[0].set_ylim(-(0.04 * ylim), ylim)
    axes[0].set_xlim(1, 10)
    axes[0].set_title('Зависимость продаж игр от оценок пользователей')

    sns.scatterplot(data=console_data, ax=axes[1], x='critic_score', y='total_sales', alpha=0.7, s=49, color=(1, 0.4, 0.1, 1))
    axes[1].set_ylim(-(0.04 * ylim), ylim)
    axes[1].set_xlim(10, 100)
    axes[1].set_title('Зависимость продаж игр от оценок критиков')
    print(platform)
    print(console_data.corr())
```

```python
plot_corr("PS4")
```

На диаграммах видно, что продажи имеют меньшую корреляцию с оценками пользователей, чем с оценками критиков. Часть игр с оценками пользователей меньше 5 имеют больше 1 млн. проданных копий, **оценки критиков более точно соответствуют продажам игр**.

Интересные результаты получены для платформы 3DS - корреляция с оценками пользователей выше, чем у остальных платформ - 0.2.

### Популярная платформа по регионам

Определим топ-5 самых популярных платформ по трем регионам: Северная Америка, Европа и Япония.

Напишем функцию которая будет строить графики продаж игр разделенные на категории и по регионам.

```python
region_name = {'na_sales':'Северная Америка', 'eu_sales':'Европа', 'jp_sales':'Япония'}

def plot_by_region(column, region):
    region_data = actual_data.groupby(column)[region].sum().sort_values(ascending=False).reset_index()
    region_data.columns = ['platform', 'sales']
    order = region_data['platform'].values
    
    plt.figure(figsize=(10, 6))

    chart = sns.barplot(data=region_data, 
                        x='platform', y='sales', 
                        palette=("rocket" if column=='platform' else "mako"),
                        order=order)
    
    chart.set_xticklabels(chart.get_xticklabels(), rotation=30)
    plt.title('{} - распределение продаж (млн. копий)'.format(region_name[region]))

    total = sum([i.get_height() for i in chart.patches])

    for i in chart.patches:
        chart.text(i.get_x()+0.15, i.get_height()+2,
                   str(round((i.get_height()/total)*100, 2))+'%', 
                   fontsize=10,
                   color=(0, 0, 0, 0.75))
```

### Влияние ESRB на продажи по регионам

Проанализируем продажи игр в зависимости от присвоенного рейтинга ESRB. Заполним пропуски в столбце *rating* на NR - "Not Rated". Составим сводную таблицу продаж, разделенных по рейтингу и постоим графики продаж в зависимости от рейтинга для каждого региона.

```python
actual_data['rating'].fillna('NR', inplace=True)
```

```python
ratings_sales = actual_data.pivot_table(index='rating',
                                        values=['na_sales', 'eu_sales', 'jp_sales'],
                                        aggfunc=['mean', 'count', 'sum'])
ratings_sales
```

Составим новый датасет с данными о продажах по значениям рейтинга в каждом регионе.

```python
def sales_by_region_rating(column):
    rating_list = ratings_sales.index.tolist()
    regions_list = ratings_sales[column].columns.tolist()

    sales = ratings_sales[column].values

    sales_raw = []
    for array in sales.tolist():
        sales_raw += (array)

    indexes = []
    for rating in rating_list:
        for region in regions_list:
            indexes.append((rating, region))

    index = pd.MultiIndex.from_tuples(indexes, names=['rating', 'region'])

    result = pd.DataFrame(data=sales_raw, index=index, columns=['sales']).reset_index()
    _ = sns.catplot(x="rating", y="sales", col="region", data=result, kind="bar", aspect=.7)
```

Постоим столбчатые диаграммы продаж игр в каждом регионе, разделенных в зависимости от рейтинга.

```python
sales_by_region_rating('mean') #Средние продажи
```

Распределение средних продаж игр в зависимости от рейтинга в Европе и Северной Америке похожи.
- Рейтинг М имеет наибольшие продажи;
- Рейтинг E, E10+ имеют меньшие значения;
- Рейтинг T показывает значительно меньшие продажи;
- Без рейтинга игры занимают последнее место.

В Японии распределение выглядит иначе.
- Без рейтинга оказались самыми продаваемыми играми;
- Рейтинги E, T - имеют меньшие продажи;
- Рейтинги M и E10+ показали самые слабые продажи.

```python
sales_by_region_rating('sum') #Сумма продаж
```

По сумме продаж распределение для Европы и Северной Америке оказалось похожим на распределение средних продаж. Выросла доля игр без рейтинга. Это объясняется большим количеством таких игр (1096 против 300 - 400 для игр с рейтингом). Для Японии суммарные продажи игр без рейтинга остались лидерами.

## Проверка гипотез

### Первая гипотеза

Сформулируем нулевую гипотезу следующим образом:
>**H0:** Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

То есть рейтинг пользователей не зависит от платформы XOne или PC. Альтернативная гипотеза будет звучать так:

>**H1:** Средние пользовательские рейтинги платформ Xbox One и PC различны.

Проведем t-тест для проверки гипотезы. Критический уровень статистической значимости установим 5%.

```python
xone_ratings = actual_data[actual_data['platform']=='XOne']['user_score'].dropna().values
pc_ratings = actual_data[actual_data['platform']=='PC']['user_score'].dropna().values

alpha = 0.05

results = st.ttest_ind(xone_ratings, pc_ratings, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
sns.boxplot(data=actual_data[actual_data['platform'].isin(["XOne", "PC"])], x='user_score', y='platform')
plt.show()
```

```python
actual_data[
    actual_data['platform'].isin(["XOne", "PC"])
].groupby('platform')['user_score'].describe()[['count', 'mean']].T.round(2)
```

Медианное и среднее значения оценок игр для PC и XOne одинаоквые. Распределение для PC имеет больший разброс и скос влево. Но отвергнуть гипотезу о равенстве оценок не получилось.

### Вторая гипотеза

Сформулируем нулевую гипотезу так:
>**H0:** Средние пользовательские рейтинги жанров Action и Sports одинаковые.

То есть рейтинг пользователей не зависит от выбора жанра Action или Sports. Альтернативная гипотеза будет звучать так:

>**H1:** Средние пользовательские рейтинги жанров Action и Sports различны.

Проведем t-тест для проверки гипотезы. Критический уровень статистической значимости установим 5%.

```python
action_ratings = actual_data[actual_data['genre']=='Action']['user_score'].dropna().values
sport_ratings = actual_data[actual_data['genre']=='Sports']['user_score'].dropna().values

alpha = 0.05

results = st.ttest_ind(action_ratings, sport_ratings, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
sns.boxplot(data=actual_data[actual_data['genre'].isin(["Action", "Sports"])], x='user_score', y='genre')
plt.show()
```

```python
actual_data[
    actual_data['genre'].isin(["Action", "Sports"])
].groupby('genre')['user_score'].describe()[['count', 'mean']].T.round(2)
```

Нулевая гипотеза была отвергнута по результатам теста. По диаграмме видно, что медианное значение для жанра Sports меньше, чем для жанра Action. Разброс нормальных значений для Sports растянулся почти на всю шкалу. Среднее значение оценки тоже сильно меньше у жанра Sports - примерно на 1.4 балла.

### Вывод

Из проведенного статистического анализа выдвинутых гипотез можно заключить:
- *Первую гипотезу о равенстве оценок пользователей для PC и XOne опровергнуть не удалось*. Оценки для PC и XOne оказались очень похожими. 


- *Вторая гипотеза об одинаковой популярности игр жанра Action и Sports была отвергнута*. По популярности жанры Action и Sports сильно различаются. Оценки для игр жанра Sports сильно меньше оценок жанра Action.
